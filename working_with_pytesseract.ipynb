```python
import pytesseract
import matplotlib.pyplot as plt
import cv2
```

```python
image = cv2.imread('notes.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image_grey = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
_, image_bw = cv2.threshold(image_grey,150, 255, cv2.THRESH_BINARY)
#image_bw = cv2.medianBlur(image_bw, 1)
plt.imshow(image_bw)
```

```python
pytesseract.pytesseract.tesseract_cmd=r'C:\Program Files\tesseract.exe'
```

```python
print(pytesseract.image_to_string(image = image_bw))
```

## OCR Using Gemini

```python
import os
from PIL import Image
```

```python
key = os.getenv("GOOGLE_API_KEY")
```

```python
import google.generativeai as genai
genai.configure(api_key=key)
```

```python
model = genai.GenerativeModel('gemini-2.5-flash-lite')
```

```python
prompt = '''You need to perform OCR on the given image and extract the text from it.
Extract text from the image and return it as plain text without any formatting or additional information.'''
```

```python
pil_img = Image.fromarray(image_bw)
pil_img
```

```python
response = model.generate_content([prompt,pil_img])
output_text = response.text
print(output_text)
```
