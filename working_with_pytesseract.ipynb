```python
import pytesseract
import matplotlib.pyplot as plt
import cv2
```

```python
image = cv2.imread(r'C:\Users\gaura\Desktop\notes.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image_grey = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
_, image_bw = cv2.threshold(image_grey, 127, 255, cv2.THRESH_BINARY)
image_bw = cv2.medianBlur(image_bw, 1)
plt.imshow(image_bw)
```

```python
pytesseract.pytesseract.tesseract_cmd=r'C:\Program Files\tesseract.exe'
```

```python
print(pytesseract.image_to_string(image = image_bw))
```

```python
cv2.imshow("Black & White", image_bw)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
